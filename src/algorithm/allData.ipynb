```python
import pandas
import numpy as np
import json
colName = ['count','month','year','time','date','RF.BKA.02_1 hr','FW.BBU.01_Q','FW.BBU.01_WL']

df = pandas.read_csv('allDataFromN.csv',usecols=[ colName[0],colName[1],colName[2],colName[3],colName[4],colName[5],colName[6],colName[7]] )
f = df.values.tolist()

with open('./allData.json') as fs:
  allData = json.load(fs)

print(f)

situation = {}
period = []
```

```python
chk_count = 2
for row in f:
    date = row[4]
    time = row[3]
    year = str(row[2])
    month = '0'+str(row[1])
    count = row[0]+1
    valueRf = row[5]
    valueQ = row[6]
    valueWl = row[7]
    if(count == chk_count):
        situation[len(situation)] = period
        period = []
        chk_count+=1
    period.append({'date':date,
                       'month':month,
                       'year':year,
                       'time':time,
                        'rfValue':valueRf,
                        'qValue':valueQ,
                        'wlValue':valueWl})
```

```python
print(situation[0],len(situation[0]),end='\n')
with open('allData.json', 'w') as fp:
    json.dump(situation, fp)
```
